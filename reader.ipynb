```python
import fitz  # PyMuPDF
import pytesseract
from PIL import Image
import io

# Path to your PDF file
pdf_path = 'mathsmaster.pdf'

# Open the PDF file
pdf_document = fitz.open(pdf_path)

# Enter the page number you want to test
page_number = 6  # example for the first page

# Check if the page number is valid
if page_number < len(pdf_document):
    # Get the specified page
    page = pdf_document[page_number]

    # Extract image from the page
    image_list = page.get_images(full=True)
    if image_list:
        base_image = image_list[0]  # Get the first image (as an example)
        xref = base_image[0]
        base_image = pdf_document.extract_image(xref)
        image_bytes = base_image["image"]

        # Load it to PIL
        image = Image.open(io.BytesIO(image_bytes))

        # Use pytesseract to do OCR on the image
        text = pytesseract.image_to_string(image)

        # Print the text
        print(text)
    else:
        print("No image found on page", page_number)
else:
    print("Page number out of range")

# Close the PDF after processing
pdf_document.close()
```
