# OCR + Output

```python
import fitz  # PyMuPDF
import pytesseract
from PIL import Image
import io
import re
import pandas as pd

# Path to your PDF file
pdf_path = 'mathsmaster.pdf'

# Open the PDF file
pdf_document = fitz.open(pdf_path)

# Define your list of page ranges here
page_ranges = [(5, 301)]

# Regex patterns
topic_pattern = re.compile(r'CHAPTER \d+ - (.*?) \[', re.DOTALL)
question_pattern = re.compile(r'(Q\.\d+.*?\?)', re.DOTALL)
date_pattern = re.compile(r'\{DSSSB.*?(\d{2}\.\d{2}\.\d{4}).*?\}', re.DOTALL)

# Placeholder for all extracted data
extracted_data = []

# Process each page range
for start_page, end_page in page_ranges:
    # Current topic placeholder
    current_topic = "Unknown Topic"
    
    # Process each page in the range
    for page_number in range(start_page, end_page + 1):
        if page_number < len(pdf_document):
            print("Processing Page Number: "+str(page_number))
            page = pdf_document[page_number]
            image_list = page.get_images(full=True)

            if image_list:
                # Extract and process the image
                base_image = image_list[0]
                xref = base_image[0]
                base_image = pdf_document.extract_image(xref)
                image_bytes = base_image["image"]
                image = Image.open(io.BytesIO(image_bytes))
                text = pytesseract.image_to_string(image)

                # Update the current topic if a new chapter is found
                topic_match = topic_pattern.search(text)
                if topic_match:
                    current_topic = topic_match.group(1).strip()

                # Find all questions and dates on the current page
                questions = question_pattern.findall(text)
                dates = date_pattern.findall(text)

                # Check if the number of questions matches the number of dates
                if len(questions) != len(dates):
                    print(f"Mismatch in number of questions and dates on page {page_number}")

                # Append data to the extracted data list
                for question, date in zip(questions, dates):
                    extracted_data.append((question, date, current_topic))
            else:
                print("No image found on page", page_number)
        else:
            print("Page number out of range")

# Close the PDF after processing
pdf_document.close()

# Create a pandas DataFrame
df = pd.DataFrame(extracted_data, columns=['Question', 'Date', 'Topic'])

# Write the DataFrame to an Excel file
excel_output_path = 'output.xlsx'
df.to_excel(excel_output_path, index=False)

print(f"Data has been written to {excel_output_path}")
```

## Calculate Weightage

```python
import pandas as pd

# Load the Excel file into a pandas DataFrame
excel_path = 'output.xlsx'  # Replace with the path to your Excel file
df = pd.read_excel(excel_path)

# Convert the 'Date' column to datetime objects and extract the year
df['Year'] = pd.to_datetime(df['Date'], errors='coerce').dt.year

# Remove rows with invalid dates
df = df.dropna(subset=['Year'])
df['Year'] = df['Year'].astype(int)

# Calculate the total number of questions per year
total_questions_per_year = df['Year'].value_counts()

# Calculate the number of questions per topic per year
topic_weightage_per_year = df.groupby(['Topic', 'Year']).size()

# Function to get the weightage of a topic in a specific year
def get_topic_weightage(topic, year):
    if year in total_questions_per_year and (topic, year) in topic_weightage_per_year:
        weightage = topic_weightage_per_year[topic, year] / total_questions_per_year[year]
        return weightage
    else:
        return 0  # Return zero weightage if the topic or year is not present

# Example usage
specified_topic = 'TIME AND WORK'  # Replace with your topic
specified_year = 2022  # Replace with your year
weightage = get_topic_weightage(specified_topic, specified_year)

print(f"The weightage of topic '{specified_topic}' in the year {specified_year} is: {weightage:.2%}")
```

## Plot Weightage VS Topic

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file into a pandas DataFrame
excel_path = 'output.xlsx'  # Replace with the path to your Excel file
df = pd.read_excel(excel_path)

# Convert the 'Date' column to datetime objects and extract the year
df['Year'] = pd.to_datetime(df['Date'], errors='coerce').dt.year

# Remove rows with invalid dates
df = df.dropna(subset=['Year'])
df['Year'] = df['Year'].astype(int)

# Calculate the total number of questions per year
total_questions_per_year = df['Year'].value_counts()

# Calculate the number of questions per topic per year
topic_weightage_per_year = df.groupby(['Topic', 'Year']).size().unstack(fill_value=0)

# Normalize the data by dividing by the total questions per year to get weightage
topic_weightage = topic_weightage_per_year.divide(total_questions_per_year, axis='columns')

# Select a specific year to plot the weightage
specified_year = 2022  # Replace with your year
weightage_data = topic_weightage[specified_year].dropna()

# Plot the data
plt.figure(figsize=(10, 8))
weightage_data.plot(kind='bar')
plt.title(f'Topic Weightage in {specified_year}')
plt.xlabel('Topic')
plt.ylabel('Weightage')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()  # Adjust the layout to fit all labels

# Save the plot to a file
plt.savefig('topic_weightage_chart.png')  # You can change the file extension to .pdf, .svg, etc.

# Show the plot
plt.show()
```
